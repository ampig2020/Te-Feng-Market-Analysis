```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ta_feng_all_months_merged.csv')
df
```

```python
df2 = df.dropna()
df2
```

```python
dfage = df2.groupby('AGE_GROUP', as_index=False)['SALES_PRICE'].sum()
dfage
```

```python
#Show Bar Chart
sns.barplot(data=dfage, x='AGE_GROUP', y='SALES_PRICE')
plt.title('Total Sales Price in term of Age Group')
plt.show()
```

```python
dfasset = df2.groupby('AGE_GROUP', as_index=False)['ASSET'].sum()
dfasset
```

```python
#Show Bar Chart
sns.barplot(data=dfasset, x='AGE_GROUP', y='ASSET')
plt.title('Total Asset in term of Age Group')
plt.show()
```
